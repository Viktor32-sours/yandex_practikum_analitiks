### Применяем фреймворк ICE для приоритизации гипотез.

ICE (от англ. impact, confidence, effort / ease «влияние, уверенность, усилия / простота») — один из самых популярных способов приоритизации задач. Метод не учитывает аудиторию охвата пользователей.

$$ICE = \frac{Impact * Confidence}{Efforts}$$

```python
hypothesis['ICE'] = hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']
hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
```

Наиболее перспективные гипотезы: 9, 1, 8

### Применим фреймворк RICE для приоритизации гипотез. 

Фреймворк RICE имеет дополнительный фактор анализа: Reach — скольких пользователей затронет изменение, которое вы хотите внести.

$$RICE = \frac{Rearch * Impact * Confidence}{Efforts}$$

```python
hypothesis['RICE'] = hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']
hypothesis[['hypothesis', 'ICE', 'RICE']].sort_values(by=['RICE','ICE'], ascending=False)
```

Самые перспективные гипотезы: 8, 3, 1. Видим изменение приориетизации. 

**Вывод.**
Приориетет гипотез при учете аудитории охвата пользователей, которых коснуться нововведения, изменился. Добавление формы подписки на все основные страницы, для сбора e-mail адресов, практически охватывает всех пользователей, и выводит с третьего на первое место гипотезу №8. Для проверки следует обратить внимание на пересечение первых троек гипотез, оцененных по этим двум фреймворкам: №8 - (Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок) и №1 - (Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей).

### График кумулятивной выручки по группам.

```python
# удаляем из данных пересекающихся пользователей
orders = orders.query('visitorId not in @cross_users')
```

```python
# создаем массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()

# получаем агрегированные кумулятивные по дням данные о заказах 
ordersAggregated = datesGroups\
.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]\
       .agg({'date' : 'max', 'group' : 'max', 'transactionId' : 'nunique', 'visitorId' : 'nunique',\
             'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])

# получаем агрегированные кумулятивные по дням данные о посетителях интернет-магазина 
visitorsAggregated = datesGroups\
.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]\
       .agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])

# объединяем кумулятивные данные в одной таблице и присваиваем ее столбцам понятные названия
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'],\
                                        right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

#print(cumulativeData.head(5)) 

# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]

# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

# Строим график выручки группы А
plt.figure(figsize=(10,5))
plt.plot(cumulativeRevenueA['date'].to_numpy(), cumulativeRevenueA['revenue'].to_numpy(), label='A')
plt.xticks(rotation=45)
plt.title(" График кумулятивной выручки по группам\n")
plt.grid(True)

# Строим график выручки группы B
plt.plot(cumulativeRevenueB['date'].to_numpy(), cumulativeRevenueB['revenue'].to_numpy(), label='B')
plt.legend()
plt.xlabel('Дата')
plt.ylabel('ВЫручка, руб.')
plt.show()
```

**Вывод.**
Наблюдается почти синхронное увеличение кумулятивной выручки до 2019-08-18, затем, 2019-08-19 скачкообразное увеличение у теста В. Далее евеличение значения теста В идет параллельно тесту А. Скачок может быть вызван единичным крупным заказом, или большим количеством покупок, вызванное акцией, проводимой в этот день.

### Расчет статистической значимость различий в конверсии между группами по «сырым» данным. 

Посчитаем статистическую значимость различий в конверсии и среднем чеке между группами по «сырым» данным — без удаления аномальных пользователей.
Сформулируем гипотезы. 
- Нулевая: различий в конверсии между группами нет. 
- Альтернативная: различия в конверсии между группами есть.

Пороговое значение определим в 5% - 0.05.

Соберем все данные по заказам и визитам (а также добавим кумулятивные показатели) в одну таблицу. Для этого создадим несколько таблиц и затем объединим.

```python
# данные по визитам группа А
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

# кумулятивные данные по визитам группа А
visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

# данные по визитам группа В
visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

# кумулятивные данные по визитам группа В
visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

# данные по заказам группа А
ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

# кумулятивные данные по заказам группа А
ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

# данные по заказам группа В
ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

# кумулятивные данные по заказам группа В
ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]

# собираем данные в одну таблицу
data = (
    ordersADaily.merge(
        ordersBDaily, on='date', how='left'
    )
    .merge(ordersACummulative, on='date', how='left')
    .merge(ordersBCummulative, on='date', how='left')
    .merge(visitorsADaily, on='date', how='left')
    .merge(visitorsBDaily, on='date', how='left')
    .merge(visitorsACummulative, on='date', how='left')
    .merge(visitorsBCummulative, on='date', how='left')
)

data.head(5)
```

Создадим переменные ordersByUsersA и ordersByUsersB со столбцами ['userId', 'orders']. В них для пользователей, которые заказывали хотя бы 1 раз, укажем число совершённых заказов.

```python
# число заказов пользователей группы А
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique}))

ordersByUsersA.columns = ['userId', 'orders']

# число заказов пользователей группы В
ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders']
```

Объявим переменные sampleA и sampleB, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. 

```python
sampleA = pd.concat([ordersByUsersA['orders'],\
                     pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum()\
                                                  - len(ordersByUsersA['orders'])), name='orders')],axis=0)

sampleB = pd.concat([ordersByUsersB['orders'],\
                     pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum()\
                                                  - len(ordersByUsersB['orders'])), name='orders')],axis=0)
```

Применим критерий и отформатируем p-value, округлив его до трёх знаков после запятой.

```python
print('p-value: ',"{0:.3f}".format(st.mannwhitneyu(sampleA, sampleB)[1]))

# отношение средней конверсии гр.В к гр.А
r = sampleB.mean() / sampleA.mean() - 1

print('Относительный выйгрыш гр.В к гр.А: ' if r > 0 else 'Относительный пройгрыш гр.В к гр.А: ',\
      "{0:.3f}".format(r))
```

**Вывод.**
По «сырым» данным, нулевую гипотезу отвергаем,  конверсии групп A и B различны, поскольку р-value < 0.05. К тому же, наблюдаем относительный выйгрыш группы В в 13.8% по количеству заказов, но делать вывод еще рано, необходимо проверить результат по очищенным данным.

### Расчет статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным.

   
  Сформулируем гипотезы. 
- Нулевая: различий в среднем чеке заказа между группами нет. 
- Альтернативная: различия в среднем чеке заказа между группами есть.

Пороговое значение определим в 5% - 0.05.



Чтобы рассчитать статистическую значимость различий в среднем чеке, передадим критерию mannwhitneyu() данные о выручке с заказов.

```python
print('p-value: ','{0:.3f}'.format(st.mannwhitneyu(orders[orders['group']=='A']['revenue'],\
                                                   orders[orders['group']=='B']['revenue'])[1]))
# отношение средних чеков гр.В к гр.А
s = orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1
print('Относительный выйгрыш гр.В к гр.А: ' if s > 0 else 'Относительный пройгрыш гр.В к гр.А: ',\
      "{0:.3f}".format(s)) 
```

**Вывод.**
Значение p-value значительно больше 0.05, поэтому отвергнуть нулевую гипотезу мы не можем и считать разными средние чеки групп А и В нет оснований. Однако, средний чек группы В значительно больше среднего чека группы А,что может быть следствием влияния выбросов, поэтому необходима проверка по *очищеным данным*.

### Расчет статистическую значимость различий в конверсии между группами по «очищенным» данным. 

Сформулируем гипотезы. 
- Нулевая: различий в конверсии между группами нет. 
- Альтернативная: различия в конверсии между группами есть.

Пороговое значение определим в 5% - 0.05.

Найденные ранее 95-й и 99-й перцентили средних чеков равны 28000 и  58233.2  рублям. А 95-й и 99-й перцентили количества заказов на одного пользователя равны 2 и 4 заказа на пользователя.
Примем за аномальных пользователей тех, кто совершил от 3 заказов или совершил заказ дороже 43 000 рублей. Так мы уберём от 1% до 5% пользователей с наибольшим числом заказов и от 1% до 5% пользователей с дорогими заказами. Сделаем срезы пользователей с числом заказов больше 3 — usersWithManyOrders и пользователей, совершивших заказы дороже 38 000 — usersWithExpensiveOrders. Объединим их в таблице abnormalUsers.
**После учета пересекающихся пользователей, граница аномальных значений изменились: количество заказов - 2, сумма заказа - 40000.**
Узнаем, сколько всего аномальных пользователей методом shape().

```python
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > 2]['userId'],
        ordersByUsersB[ordersByUsersB['orders'] > 2]['userId'],
    ],
    axis=0
)
usersWithExpensiveOrders = orders[orders['revenue'] > 40000]['transactionId']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)

print(abnormalUsers.head(5))
print('Количество "анормальных" пользователей: ', abnormalUsers.shape[0]) 
```

Узнаем, как их действия повлияли на результаты теста. Посчитаем статистическую значимость различий в конверсии между группами теста по очищенным данным. Сначала подготовим выборки количества заказов по пользователям по группам теста:

```python
# отфильтрованные данные группа А
sampleAFiltered = pd.concat(
    [ordersByUsersA[np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))]['orders'],\
        pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])),\
                  name='orders')] ,axis=0)

# отфильтрованные данные группа В
sampleBFiltered = pd.concat(
    [ordersByUsersB[np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))]['orders'],\
     pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])),\
               name='orders')], axis=0)
```

Применим статистический критерий Манна-Уитни к полученным выборкам:


```python
print('p-value: ','{0:.3f}'.format(st.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
r = sampleBFiltered.mean()/sampleAFiltered.mean()-1

print('Относительный выйгрыш гр.В к гр.А: ' if s > 0 else 'Относительный пройгрыш гр.В к гр.А: ',\
      "{0:.3f}".format(r)) 
```

**Вывод**
Результат по конверсии на *очищеных данных*, практически не изменился: нулевую гипотезу отвергаем, конверсии групп A и B различны, поскольку р-value < 0.05. Относительный выйгрыш увеличился на 1.1%. Посмотрим, как изменились показатели по средним чекам.

### Расчет статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным.

 Сформулируем гипотезы.

- Альтернативная: различия в среднем чеке заказа между группами есть.

Пороговое значение определим в 5% - 0.05.
