### Шаг 2.6. Категоризация дохода.

Разделим клиентов по категориям исходя из суммы ежемесячного дохода `total_income`. Для этого создадим столбец `total_income_category` и поместим в него категории в соответствии с правилами:
* 0–30000 — 'E';
* 30001–50000 — 'D';
* 50001–200000 — 'C';
* 200001–1000000 — 'B';
* 1000001  и выше — 'A'.

Чтобы разбить на категории напишем функцию.

```python
def total_income_category(total_income):
    """
    Возвращает группу по значению дохода total_income, используя правила:
    * 0–30000 — 'E';
    * 30001–50000 — 'D';
    * 50001–200000 — 'C';
    * 200001–1000000 — 'B';
    * 1000001 и выше — 'A'.
    """
    try:
        if total_income <= 30000:
            return 'E'
        if total_income <= 50000:
            return 'D'
        if total_income <= 200000:
            return 'C'
        if total_income <= 1000000:
            return 'B'
        return 'A' 
    except:
        return 'Invalid data'
```

Протестируем функцию:

```python
income_list = [10, 40000, 200000, 990000, 2000000, '#']
for income in income_list:
    print(f'Доход {income} - группа {total_income_category(income)}')
```

Функция работает правильно, создаем и заполняем столбец

```python
data['total_income_category'] = data['total_income'].apply(total_income_category)
data['total_income_category'].value_counts()
```

### Вывод
Разбивка клиентов по категориям прошла успешно. 
Видно, что самая многочисленная группа - клиенты с доходом от 50000 до 200000, а самые малочисленная с доходом до 30000 и со сверхвысоким дододом от 1000001. 

### Шаг 2.7. Категоризация целей кредита.

Посмотрим внимательно на цели кредита

```python
data['purpose'].value_counts
```

Можно выделить 4 основные категории:
- 'операции с автомобилем',
- 'операции с недвижимостью',
- 'проведение свадьбы',
- 'получение образования'.

Создадим новый столбец `purpose_category`, и напишем функцию для его заполнения.


```python
def purpose_category(purpose):
    """
    Возвращает группу в зависимости от наличия в строках целей кредита purpose ключевых подстрок.
    * 'автом' — 'операции с автомобилем';
    * 'жил', 'недвиж'— 'операции с недвижимостью';
    * 'свад' — 'проведение свадьбы';
    * 'образ' — 'получение образования';
    """   
    try:
        if 'автом' in purpose:
            return 'операции с автомобилем'
        if 'жил' in purpose or 'недвиж' in purpose:
            return 'операции с недвижимостью'
        if 'свад' in purpose:
            return 'проведение свадьбы'
        if 'образ' in purpose:
            return 'получение образования'
        return 'unknown categoty'
    except:
        return 'Invalid data'
```

Протестируем функцию

```python
test_list = list(data['purpose'].unique())[:-20]
test_list
```

```python
for string in test_list:
    print(string, "==>", purpose_category(string))
```

Функция работает правильно, создаем и заполняем новый столбец:

```python
data['purpose_category'] = data['purpose'].apply(purpose_category)
data['purpose_category'].value_counts()
```
